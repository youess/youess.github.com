Pillow 是python用来处理图片的，本文记录一下其主要的功能。



```python
# 导入PIL中的图像类
from PIL import Image

# 让图片能在ipynb里面显示
import numpy as np
from matplotlib.pyplot import imshow
%matplotlib inline

def show(img):
    imshow(np.asanyarray(img))
```

```python
# 读取图片
im = Image.open('../../assets/img/cat.jpg')
```

```python
# 图片模式，类型，以及(宽，长)
print(im.mode, im.format, im.size)
```

```python
im.show()         # nothing
# imshow(np.asanyarray(im))
show(im)
```

```python
# 创建缩略图
size = (64, 64)
im.thumbnail(size)
show(im)
```

```python
# 选取图中区域
box = (50, 50, 100, 100)
region = im.crop(box)
show(region)
```
